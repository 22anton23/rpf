–£–±–µ—Ä–µ–º –ø–∞—Ä—Ç–∏–∏ –≥–¥–µ —ç—Ç–∞–ø–æ–≤ –º–µ–Ω—å—à–µ —á–µ–º 2

````python
first_temp_indexs = []
last_temp_indexs = []

for i in temp_f['key'].unique():
    key_data = temp_f.query('key == @i')
    first_temp_indexs.append(key_data.index[0])
    last_temp_indexs.append(key_data.index[-1])
````

–†–∞—Å—Å—á–∏—Ç–∞–µ–º –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∫–∞–∂–¥–æ–π –ø–∞—Ä—Ç–∏–∏

````python
arc['–ù–∞—á–∞–ª–æ –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'] = pd.to_datetime(arc['–ù–∞—á–∞–ª–æ –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'])
arc['–ö–æ–Ω–µ—Ü –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'] = pd.to_datetime(arc['–ö–æ–Ω–µ—Ü –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'])
arc['–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã'] = arc['–ö–æ–Ω–µ—Ü –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'] - arc['–ù–∞—á–∞–ª–æ –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π']
arc['–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã'] = arc['–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã'].dt.seconds
````

## –°–æ–∑–¥–∞–Ω–∏—è –æ–±—â–µ–π —Ç–∞–±–ª–∏—Ü—ã

## –ü–æ–∏—Å–∫ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏

–†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö target –∏ features

````python
target = full_df['–§–∏–Ω–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞']
features = full_df.drop(['–§–∏–Ω–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞'], axis=1)
````

–†–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ train –∏ test 3 –∫ 1

````python
features_train, features_test, target_train, target_test = train_test_split(
     features, target, test_size=0.25, random_state=240423)
````

````python
scaler = MinMaxScaler()
scaler.fit(features_train)
features_train = scaler.transform(features_train)
features_test = scaler.transform(features_test)
````

–í—ã–±–æ—Ä–∫–∏ –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω—ã –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ

–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–∏—Å–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ —Å –ø–æ–º–æ—â—å—é GridSearchCV

````python
def find_best_model(params, model, features, target):
    grid = GridSearchCV(estimator=model,
                         param_grid=params,
                         scoring='neg_mean_absolute_error',
                         cv=5,
                         n_jobs=-1)
    
    grid.fit(features, target)

    return grid
````
